# <span style="color:blue">Le jeu de données : prix-mobile.csv </span>

Dans ce Notebook, nous allons développer un modèle KNN sur le jeu de données 'prix-mobile.csv'.  

### 1. Informations sur le jeu de données "prix-mobile.csv"

Le jeu de données comprend diverses caractéristiques qui fournissent des informations sur les smartphones ainsi que la classe de prix.

Le jeu de données contient un ensemble d'observations étiquetées avec la variable de classe  <span style="color:blue"> price_range </span>, qui permet de classer les smartphones selon leurs prix en quatre catégories : "0", "1", "2", "3".


### 2. Détails sur quelques les caractéristiques

- **battery_power** : Énergie totale qu'une batterie peut stocker en une seule fois, mesurée en mAh.
- **blue** : posséde un Bluetooth ou pas.
- **clock_speed** : La vitesse à laquelle le microprocesseur exécute les instructions.
- **dual_sim** : Prend en charge la double carte SIM ou non.
- **fc** : Mégapixels de la caméra frontale.
- **four_g** : A un réseau 4G ou pas.
- **int_memory** : Mémoire interne en gigaoctets.
- **m_dep** : Hauteur du mobile en cm.
- **mobile_wt** : Poids du téléphone.
- **n_cores** : Nombre du processeur.
- **pc** : Mégapixels de l'appareil photo principal.
- **px_height** : Hauteur de résolution en pixels.
- **px_width** : Largeur de résolution en pixels.
- **ram** : Mémoire vive en mégaoctets.
- **sc_h** : Hauteur de l'écran du mobile en cm.
- **sc_w** : Largeur de l'écran du mobile en cm.
- **talk_time** : la durée la plus longue pendant laquelle une seule charge de batterie durera.
- **three_g** : A un réseau 3G ou pas.
- **touch_screen** : Avec écran tactile ou non
- **wifi** : Dispose d'une connexion Wi-Fi ou non.
- **price_range** : Il s'agit de la variable cible avec une valeur de 0 (faible coût), 1 (coût moyen), 2 (coût élevé) et 3 (coût très élevé).

## 14. Création, Entrainement et Evaluation d'un KNN  

### 14.1. Répartir la base en deux tableaux : <span style="color:green">(1,5 pts) </span>
  - **X** : Contient toutes les données à l'exception de la colonne dont le nom est **price_range**.  
  - **y** : Contient la colonne dont le nom est **price_range**, utilisée comme target.  


Affichez les dimensions des deux tableaux pour vérifier leur structure.

```python
X=ds.drop('price_range',axis=1)
```

```python
y=ds['price_range']
```

```python
print(X.shape) print(y.shape)
```

### 14.2. Répartir la base en deux ensembles : ensemble d'apprentissage et ensemble de test en fixant l'aléatoire à 0. <span style="color:green">(1 pt) </span>

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

### 14.3. Afficher la forme des 4 sous ensembles ( X_train, X_test, y_train et y_test ) <span style="color:green">(1 pt) </span>

```python
print("x_train",X_train.shape)
```

### 14.4. Créer un modèle KNN avec un nombre de voisins = 5  <span style="color:green">(0,5 pt) </span>

```python
KNN=KNeighborsClassifier(n_neighbors=5)
```

### 14.5. Entrainer et valider le modèle <span style="color:green">(1 pt) </span>

```python
KNN.fit(X_train,y_train)
```

## 15. Évaluation du modèle

### 15.1 calculer y_pred <span style="color:green">(1 pt) </span>

```python
y_pred=KNN.predict(X_test)
```

### 15.2 Afficher le score d'apprentissage et le score du test <span style="color:green">(1 pt) </span>

```python
print(KNN.score(X_test,y_test))
```

### 15.3 Evaluer les performances du modèle via un rapport de classification. <span style="color:green">(1 pt) </span>

```python
print(classification_report(y_test,y_pred))
```
